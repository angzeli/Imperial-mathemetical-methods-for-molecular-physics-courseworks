```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define F(p) as a function of p
def F(p):
    integrand = lambda x: np.exp(-p * x**2) / (1 + x)
    result, _ = quad(integrand, 0, np.inf)
    return result

# Approximate expression for large p
def F_approx(p):
    return np.sqrt(np.pi)/(2*np.sqrt(p)) - 1/(2*p)

# Create range of p values
p_values = np.linspace(0.1, 10, 200)
F_values = [F(p) for p in p_values]
F_approx_values = [F_approx(p) for p in p_values]

# Plot comparison
plt.figure(figsize=(8,6))
plt.plot(p_values, F_values, label=r'$F(p)=\int^\infty_0\frac{\exp(-px^2)}{1+x}\ dx$', lw=2)
plt.plot(p_values, F_approx_values, '--', label='Approximated $F_{approx}(p)$', color='r', lw=2)
plt.xlabel('$p$', fontsize = 22, fontweight = 'bold')
plt.ylabel('$F(p)$', fontsize = 22, fontweight = 'bold')
plt.xticks(fontsize = 14, fontweight = 'bold')
plt.yticks(fontsize = 14, fontweight = 'bold')
#plt.title('Comparison of numerical and approximate $F(p)$',fontsize = 16, fontweight = 'bold')
plt.legend(prop = {'size': '12','weight': 'bold'})
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(1.8)
#plt.grid(True)
plt.tight_layout()
plt.savefig('Fig 4.1 Comparison of numerical and approximate F(p).png', dpi=1200)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define f(p,1) =  sin(x)/x dx
def f(p):
    integrand = lambda x: np.sin(x)/x if x != 0 else 1  # handle x=0 safely
    result, _ = quad(integrand, 0, p)
    return result

# Generate values of p
p_values = np.linspace(0, 5, 300)
f_values = [f(p) for p in p_values]

# Plot
plt.figure(figsize=(8,6))
plt.plot(p_values, f_values, label=r'$f(p,1)=\int_0^p \frac{\sin x}{x}\,dx$', lw=2)
plt.plot(p_values, p_values-p_values**3/18, label=r'Approximated $f_{approx}(p)$', color='r', linestyle='--', lw=2)
plt.xlabel('$p$', fontsize = 22, fontweight = 'bold')
plt.ylabel('$f(p)$', fontsize = 22, fontweight = 'bold')
plt.xticks(fontsize = 14, fontweight = 'bold')
plt.yticks(fontsize = 14, fontweight = 'bold')
#plt.ylim(1,1.7)
#plt.title('Numerical plot of $f(p,1)$', fontsize=13)
#plt.grid(True)
plt.legend(prop = {'size': '12','weight': 'bold'}, loc='lower left')
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(1.8)
plt.tight_layout()
plt.savefig('Fig 4.2 Comparison of numerical and approximate f(p).png', dpi=1200)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ===== Define functions =====
def f_exp_neg(x, alpha): 
    return np.exp(-alpha * x**2)

def f_exp_neg_approx(x, alpha): 
    return 1 - alpha * x**2 + 0.5 * (alpha * x**2)**2

def f_log(x, alpha): 
    return np.log(1 + alpha * x**2)

def f_log_approx(x, alpha): 
    return alpha * x**2 - 0.5 * (alpha * x**2)**2 + (1/3) * (alpha * x**2)**3

def f_root(x, alpha, mu): 
    return (1 + alpha * np.sqrt(x))**mu

def f_root_approx(x, alpha, mu): 
    return 1 + mu * alpha * np.sqrt(x) + 0.5 * mu * (mu - 1) * (alpha**2) * x

# ===== Domain settings =====
x1 = np.linspace(0, 1.5, 500)
x2 = np.linspace(0, 1, 500)

# ===== Set up figure =====
fig, axes = plt.subplots(2, 4, figsize=(14, 6))

# ===== Define subplot configurations =====
plots = [
    # Row 1: e^{-x},  = 1
    {"ax": axes[0,0], "func": f_exp_neg, "approx": f_exp_neg_approx, "x": x1, "alpha": 1, "title": r"(a) $e^{-\alpha x^2}$"},
    {"ax": axes[0,1], "func": f_exp_neg, "approx": f_exp_neg_approx, "x": x1, "alpha": -1, "title": r"(a) $e^{-\alpha x^2}$"},
    
    # Row 1 cont'd: ln(1+x),  = 1
    {"ax": axes[0,2], "func": f_log, "approx": f_log_approx, "x": x1, "alpha": 1, "title": r"(b) $\ln(1+\alpha x^2)$"},
    {"ax": axes[0,3], "func": f_log, "approx": f_log_approx, "x": x1, "alpha": -1, "title": r"(b) $\ln(1+\alpha x^2)$"},
    
    # Row 2: (1+x)^,  = 1,  = 0.5, 2
    {"ax": axes[1,0], "func": f_root, "approx": f_root_approx, "x": x2, "alpha": 1, "mu": 0.5, "title": r"(c) $(1+\alpha\sqrt{x})^{\mu}$"},
    {"ax": axes[1,1], "func": f_root, "approx": f_root_approx, "x": x2, "alpha": -1, "mu": 0.5, "title": r"(c) $(1+\alpha\sqrt{x})^{\mu}$"},
    {"ax": axes[1,2], "func": f_root, "approx": f_root_approx, "x": x2, "alpha": 1, "mu": 2, "title": r"(c) $(1+\alpha\sqrt{x})^{\mu}$"},
    {"ax": axes[1,3], "func": f_root, "approx": f_root_approx, "x": x2, "alpha": -1, "mu": 2, "title": r"(c) $(1+\alpha\sqrt{x})^{\mu}$"},
]

# ===== Plot loop =====
for cfg in plots:
    ax = cfg["ax"]
    x = cfg["x"]
    alpha = cfg.get("alpha", 1)
    mu = cfg.get("mu", None)

    if mu is not None:
        y = cfg["func"](x, alpha, mu)
        y_approx = cfg["approx"](x, alpha, mu)
    else:
        y = cfg["func"](x, alpha)
        y_approx = cfg["approx"](x, alpha)

    ax.plot(x, y, 'k', label='Original', lw=1.8)
    ax.plot(x, y_approx, 'r--', label='Approx.', lw=1.8)
    ax.set_title(cfg["title"], fontsize=11, fontweight='bold')
    ax.set_xlabel('x', fontsize=10, fontweight='bold')
    ax.set_ylabel('f(x)', fontsize=10, fontweight='bold')
    #ax.grid(True, alpha=0.3)
    
    # Bold tick labels
    ax.tick_params(axis='both', labelsize=9, width=1.5)
    for tick in ax.get_xticklabels() + ax.get_yticklabels():
        tick.set_fontweight('bold')

    # Consistent spine thickness
    for spine in ax.spines.values():
        spine.set_linewidth(1.8)

    # ===== Automatic  and  annotation =====
    parts = [rf"$\alpha={alpha}$"]
    if mu is not None:
        parts.append(rf"$\mu={mu}$")
    param_text = ", ".join(parts)

    ax.text(
        0.03, 0.97, param_text,  # top-left position
        transform=ax.transAxes, ha="left", va="top",
        fontsize=10, fontweight="bold",
        bbox=dict(boxstyle="round,pad=0.25", facecolor="white", alpha=0.8, edgecolor="none")
    )

# ===== Shared legend =====
handles, labels = axes[0,0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', ncol=2, fontsize=12, frameon=False)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig("Figure_1.2_All_Functions_with_Params.png", dpi=1200)
plt.show()
```
